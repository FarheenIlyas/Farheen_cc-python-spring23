```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt  #image plotting
%matplotlib inline
import os #For interacting with Operating System - we use it for going through files
import glob #For file name pattern matching
```

```python
#Black & White Imgaes

img_dir = "ImageManipulation" # Enter Directory of all images 
data_path = os.path.join(img_dir,'*.jpg')
files = glob.glob(data_path)
data = []
for f1 in files:
    img = cv2.imread(f1)
    resized = cv2.resize (img, (500,500))
    gray_img = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
    data.append(gray_img)
```

```python
for gray_img in data:
    plt.imshow(gray_img, cmap='gray')
    plt.show()
```

```python
for x in range(len(data)):
    d = x
    filename = f"output/grayimg_{d}.jpg"
    cv2.imwrite(filename, data[x])
    d += 1
```

```python
#Negative Images:

for img in data:
    neg_img = 255 - img  # Subtracting each pixel value from 255 to obtain negative image
    plt.figure()
    plt.imshow(neg_img, cmap='gray')
    plt.show()
```

```python
for x in range(len(data)):
    neg_img = 255 - data[x] 
    filename = f"output/negimg_{x}.jpg" 
    cv2.imwrite(filename, neg_img)
```
