```python
import matplotlib.pyplot as plt
import numpy as np

Days = ("Wed", "Thur", "Fri", "Sat", "Sun", "Mon", "Tue")
Emotions = {
    'Happy': (35,50,1,20,35, 40, 30),
    'Okayish': (50,30,1,10,25, 30, 25),
    'Sad': (10,15,100,30,25, 20, 35),
    'Screw_it!': (5,5,1,50,15, 10, 10)
}

x = np.arange(len(Days))  # the label locations
width = 0.2  # the width of the bars
multiplier = 1
plt.rcParams["figure.figsize"] = [9, 6] # adjusting the width of the graph

colors = {
    'Happy': 'yellow',
    'Okayish': 'rosybrown',
    'Sad': 'lightblue',
    'Screw_it!': 'red'
}

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in Emotions.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label= attribute, color=colors[attribute])
    multiplier += 1

ax.set_ylabel('Intensity')
ax.set_title('Emotional Rollercoaster')
ax.set_xticks(x + 0.5, Days)
ax.legend(loc='upper left', ncol = 4)
ax.set_ylim(0, 120)

plt.show()
```
